```python
import numpy as np
```

```python
def blsys(B, C):
    def f(x, u):
        return B @ x + C @ u
    return f
```

```python
def cd(a):
    def f(k, l):
        return k ** a * l ** (1 - a)
    return f

cd(0.3)(2, 1)
```

```python
from numpy.random import randn

def simulate(g, x0, T):
    x = np.empty((T, len(x0), 1))
    x[0] = x0.reshape((2, 1))
    for i in range(T-1):
        u = randn(2, 1)
        x[i+1] = g(x[0], u)
    return x
```
